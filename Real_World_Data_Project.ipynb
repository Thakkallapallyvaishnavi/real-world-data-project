```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import yfinance as yf
```

```python
data = yf.download("AAPL", start="2020-01-01", end="2023-01-01")
data.head()
```

```python
data.info()
data.describe()
data.isnull().sum()
```

```python
plt.figure(figsize=(10,5))
plt.plot(data['Close'])
plt.title("Closing Price of Stock")
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot(data['Volume'])
plt.title("Stock Volume")
plt.show()
```

```python
data['MA50'] = data['Close'].rolling(50).mean()

plt.figure(figsize=(10,5))
plt.plot(data['Close'], label="Close Price")
plt.plot(data['MA50'], label="Moving Average")
plt.legend()
plt.title("Moving Average")
plt.show()
```

Insights

1. The stock price shows an increasing trend over time.
2. There are fluctuations in price indicating market changes.
3. Moving average smooths the data and shows trend clearly.
4. High volume indicates active trading periods.
